```python
from temperaSimulada import cria_cidades, gera_solucao, calcula_custo, temperaSimulada
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_cidades = 10
print("Matrix de Cidades:")
cidades = cria_cidades(n_cidades)
si = gera_solucao(n_cidades)
print(f"Solucao inicial: {si}")
print(f"Custo inicial: {calcula_custo(cidades, si)}")
```

```python
for i in range(5):
    st = temperaSimulada(cidades, si, max_iter=1000)
    sf = st[-1]
    print(f"Soluçao encontrada: {sf}")
    print(f"Custo final: {calcula_custo(cidades, sf)}")

    x = np.arange(0, len(st))
    y = np.array([])

    for s in st:
        y = np.append(y, calcula_custo(cidades, s))

    plt.title("Custo x t") 
    plt.xlabel("t") 
    plt.ylabel("Custo") 
    plt.plot(x, y, color ="red") 
    plt.show()
```
